```python
import pandas as pd
```

```python
data = {
    'Français': ["Bonjour", "Salut", "Comment tu t'appelles ?", "Je m'appelle", "Comment ça va ?", "Ça va bien", "Ça va mal", "Je suis désolé", "Merci", "De rien", "S'il vous plaît", "Oui", "Non", "Excusez-moi", "Au revoir"],
    'Espagnol': [None] * 15,
    'Anglais': [None] * 15,
    'Italian': [None] * 15,
}
```

```python
dataframe = pd.DataFrame(data)
dataframe 
```

```python
import google.generativeai as genai
import os
from dotenv import load_dotenv

load_dotenv()


import time

def translator(prompt, source_language, target_language, delay=4):
    API_KEY = os.getenv("API_KEY")  


    genai.configure(api_key=API_KEY)


    translation_prompt = f"Translate this text from {source_language} to {target_language}: {prompt}."


    try:
        model = genai.GenerativeModel("gemini-1.5-flash")
        response = model.generate_content(translation_prompt)
        time.sleep(delay)  
        return response.text.strip() 
    except Exception as e:
        return f"An error occurred: {str(e)}"
```

```python
translator("Bonjour, comment ça va ? ", "Français", "Anglais")
```

```python

def translate_column(language):
    dataframe[language] = dataframe['Français'].apply(lambda x: translator(x, "Français", language.capitalize(), delay=4))

languages = ["Anglais", "Espagnol", "Italian"]
for lang in languages:
    translate_column(lang)
```

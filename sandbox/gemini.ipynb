```python
import pandas as pd
```

```python
data = {
    'Français': ["Bonjour", "Salut", "Comment tu t'appelles ?", "Je m'appelle", "Comment ça va ?", "Ça va bien", "Ça va mal", "Je suis désolé", "Merci", "De rien", "S'il vous plaît", "Oui", "Non", "Excusez-moi", "Au revoir"],
    'Espagnol': [None] * 15,
    'Anglais': [None] * 15,
    'Italian': [None] * 15,
}
```

```python
dataframe = pd.DataFrame(data)
dataframe 
```

```python
import google.generativeai as genai
import os
from dotenv import load_dotenv

# Charger les variables d'environnement depuis le fichier .env
load_dotenv()

# Fonction de traduction utilisant l'API Generative AI
import time

# Fonction de traduction avec délai pour éviter l'erreur 429
def translator(prompt, source_language, target_language, delay=3):
    # Clé API pour l'accès à Generative AI
    API_KEY = os.getenv("API_KEY")  # Récupère la clé API depuis .env

    # Configurer l'API Generative AI
    genai.configure(api_key=API_KEY)

    # Préparer le prompt pour la traduction
    translation_prompt = f"Translate this text from {source_language} to {target_language}: {prompt}."

    # Utiliser le modèle Generative AI pour générer la traduction
    try:
        model = genai.GenerativeModel("gemini-1.5-flash")
        response = model.generate_content(translation_prompt)
        time.sleep(delay)  # Attendre avant de faire la prochaine requête
        return response.text.strip()  # Assurez-vous de supprimer les espaces supplémentaires
    except Exception as e:
        return f"An error occurred: {str(e)}"
```

```python
translator("Bonjour, comment ça va ? ", "Français", "Anglais")
```

```python
# Appliquer les traductions aux colonnes
def translate_column(language):
    dataframe[language] = dataframe['Français'].apply(lambda x: translator(x, "Français", language.capitalize(), delay=3))

# Traductions pour chaque langue
languages = ["Anglais", "Espagnol", "Italian"]
for lang in languages:
    translate_column(lang)
```
